```python
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
import sklearn
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from plotnine import *
```

```python
def kmeans_plot(data_set , n0_components , rand_state ):
    kmeans = KMeans(n_clusters=n0_components, random_state=rand_state)
    kmeans.fit(data_set[["x", "y"]])
    
    pred = kmeans.predict(data_set[["x", "y"]])
    plt.scatter(data_set["x"] , data_set["y"] , c = pred)
    plt.title("Kmeans Clustering")
    plt.show()
        
    
```

```python
kmeans_plot(dataset1 , 2 , 0 )
```

```python
def data_plot(dataset):
    plt.scatter(dataset["x"] , dataset["y"])
    plt.title("original")
```

```python
def cluster(data_set , no_cluster , rand_state):
    plt.figure(figsize=(50,15))
    plt.subplot(1,2,1)
    data_plot(data_set)
    plt.subplot(1,2,2)
    kmeans_plot(data_set , no_cluster , rand_state )
    plt.show()
    
    
```

```python
cluster(dataset1 , 2 , 0)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(dataset1[["x", "y"]])
means = kmeans.cluster_centers_
pred = kmeans.predict(dataset1[["x", "y"]])

plt.scatter(dataset1["x"] , dataset1["y"] , c = pred)
print(means )
plt.title("KMEANS CLUSTERING")
plt.show()
```

```python
def dbscan(data_set, eps, min_samples):
    db = DBSCAN(eps=eps, min_samples=min_samples)
    pred = db.fit_predict(data_set[["x", "y"]])

    plt.scatter(data_set["x"], data_set["y"], c=pred)
    plt.title("dbscan")
    plt.show()
```

```python
dbscan(dataset1,0.5,7)
```
